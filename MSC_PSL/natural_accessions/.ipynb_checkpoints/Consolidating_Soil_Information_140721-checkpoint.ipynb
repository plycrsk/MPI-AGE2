```python
## will combine accession dataset with harmonised soil dataset ##

import pandas as pd
import numpy as np
```

```python
data_a = pd.read_csv("At_natural_accessions_reference_with_SMU.csv")
data_soil = pd.read_csv("HWSD_dataframe.csv")
```

```python
data_a = data_a.dropna()
```

```python
data_soil = data_soil.drop(columns=['Unnamed: 0', 'ID'])
```

```python
data_soil = data_soil.set_index('MU.GLOBAL')
```

```python
soil_dict = data_soil.to_dict()
```

```python
test_data = data_a
```

```python
test_data = test_data.apply(pd.to_numeric, errors='ignore')
```

```python
for key in soil_dict:
    test_data[key] = np.nan
    test_data[key] = test_data['Soil Mapping Unit'].map(soil_dict[key])
```
