```python
import pandas as pd
```

```python
RNAseq_data = pd.read_csv("../Results/DiffExp_Analysis/DiffExp_age_plus_div_Q4.00_150821.csv")
Markers = pd.read_csv("Inflammatory_Markers_260721.csv", header=None)
ortho = pd.read_csv("killifish_human_ortho.csv")
```

```python
ort_dict = dict(zip(ortho['Gene.stable.ID'], ortho['Gene.stable.ID.1']))
```

```python
RNAseq_data['human_ortho'] = RNAseq_data['Unnamed: 0'].map(ort_dict)
```

```python
RNAseq_data = RNAseq_data.dropna()
```

```python
Markers = Markers.dropna()
```

```python
gene_dict = dict(zip(Markers[2], Markers[3]))
```

```python
RNAseq_data = RNAseq_data[RNAseq_data['human_ortho'].isin(Markers[2])]
```

```python
RNAseq_data = RNAseq_data.reset_index()
```

```python
RNAseq_data['Gene_Name'] = RNAseq_data['Unnamed: 0'].map(gene_dict)
```
